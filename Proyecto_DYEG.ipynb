```python
import numpy as np
import matplotlib.pyplot as plt
import csv
import pandas as pd
import matplotlib.colors as mcolors
```

```python
#---Dataset
with open("/Users/karytzamontiel/Documents/Maestria_UNAM/Dinamica y estructura galaxias/proyecto2/The10pcSample_v2.csv", newline='') as csvfile:
    reader = csv.reader(csvfile)
    data = np.array(list(reader))
    
print(data)
```

```python
#---Abriendo columnas específicas
columns_to_open = [23, 32, 19, 5,6,8,9,11,13,27,28,29,3]

#---Nombrar columnas
column_names = ["G", "J", "Type", "RA","DEC","Parallax","Parallax_e","PMRA","PMDEC", "U", "B", "V","Obj"]

#---Verifica si las columnas especificadas existen en los datos
selected_columns = []
for col in columns_to_open:
    if col < data.shape[1]:
        selected_columns.append(data[:, col])
        
#---Eliminar valores no numéricos
cleaned_columns = []
for col_data in selected_columns:
    cleaned_col = pd.to_numeric(col_data, errors='coerce')
    cleaned_columns.append(cleaned_col)

#---Diccionario para asociar los nombres de las columnas con los datos limpios
column_data = {name: col_data for name, col_data in zip(column_names, cleaned_columns)}

#---Resta entre las columnas "G" y "J"
result = np.array(column_data["G"], dtype=float) - np.array(column_data["J"], dtype=float)

#---Guardar el resultado en otro array
resta_GJ = np.array(result)

#---Resultado
print("Resultado de la resta G - J:")
print("---------------")
print(resta_GJ)
```

```python
#---Tipos espectrales
Type1 = selected_columns[2]
print("Tipos espectrales:")
print("--------------")
print(Type1)
```

```python
#---Graficando
from matplotlib.lines import Line2D

type_categories, type_codes = np.unique(Type1, return_inverse=True)

plt.figure(figsize=(6, 6))

#---Normalizar los colores para la gráfica de dispersión y la colorbar
norm_scatter = mcolors.Normalize(vmin=resta_GJ.min(), vmax=resta_GJ.max())
norm_colorbar = mcolors.Normalize(vmin=type_codes.min(), vmax=type_codes.max())

cmap = 'plasma'

#---Puntos de color según el tipo espectral
scatter = plt.scatter(resta_GJ, column_data["G"], c=type_codes, cmap=cmap)

#---Invertir el eje Y para que las magnitudes más brillantes estén en la parte superior
#plt.gca().invert_yaxis()

plt.xlabel("$G - J$ $[mag]$")
plt.ylabel("$M_G$ $[mag]$")
plt.title("Diagrama Color-Magnitud")
plt.xlim(0, 5.8)
plt.ylim(23,0)

#---Crear la colorbar
norm = mcolors.Normalize(vmin=type_codes.min(), vmax=type_codes.max())
sm = plt.cm.ScalarMappable(cmap=cmap, norm=norm)
sm.set_array([])

cbar = plt.colorbar(sm, orientation='vertical')
cbar.set_ticklabels(["", 'T0', 'L0', 'M0', 'K0', 'G0', 'F0'])
cbar.set_label('Spectral type')

#---Leyendas
legend_elements = [Line2D([0],[0], marker="o", color="orange", label="10 pc with spectral type")]
legend_elements1 = [Line2D([0],[0], marker="o", color='#800080', label="10 pc white dwarfs")]

#---Agregar las leyendas a la gráfica
legend = plt.legend(handles=legend_elements, loc='upper left',bbox_to_anchor=(0.01, 0.15))
legend1 = plt.legend(handles=legend_elements1, loc='upper left',bbox_to_anchor=(0.01, 0.1))

#---Mostrar ambas leyendas
plt.gca().add_artist(legend)
plt.gca().add_artist(legend1)

plt.tight_layout()
plt.savefig("Diagrama Color-Magnitud.png")
plt.show()
```

Se necesita transformar las coordenadas RA y DEC a coordenadas heliocéntricas.

UNIDADES:
[RA]: [grados],
[DEC]: [grados],
[Parallax]: [mas]

```python
Parallax = np.array(Parallax, dtype=float)

#---Calculando distancia r
r = []
#r = 1/(Parallax*np.pi/(180*3600*1000)) #paralaje en radianes
r = 1/(Parallax/1000) #paralaje en pc
r = np.array(r) #distancia en pc
```

```python
#print(r)
```

```python
DEC = np.array(DEC, dtype=float)

#----Calculando la latitud heliocentrica beta
beta = []
beta = DEC*np.pi/180 
beta = np.array(beta) #radianes
```

```python
#print(beta)
```

```python
RA = np.array(RA, dtype=float)


#---Calculando la longitud heliocentrica lambda
lamb = []
lamb = RA*np.pi/180 
lamb = np.array(lamb) #radianes
```

```python
#print(lamb)
```

```python
#---Aplicando las ecuaciones de transformacion
X = []
X = r*np.cos(beta)*np.cos(lamb)
X = np.array(X) #UA

Y = []
Y = r*np.cos(beta)*np.sin(lamb)
Y = np.array(Y) #UA

#print(X)
#print(Y)

#---Convirtiendo de UA a pc
#X_pc = []
#X_pc = X*206264.806245376
#X_pc = np.array(X_pc)

#Y_pc = []
#Y_pc = Y*206264.806245376
#Y_pc = np.array(Y_pc)
```
